# Mini tutorial Git

### Condividi il progetto su GitHub

1. *Crea un account*
   Vai su [github.com](https://github.com), clicca *Sign up* e segui i passaggi base (nome utente, e-mail, password).

2. *Inizializza il repo nella cartella del progetto*
   Apri il terminale nella cartella del tuo progetto e digita:

   bash
   git init
   echo ".venv/" > .gitignore      # esclude la venv dal repository
   git add *.py *.ipynb README.md mnist_cnn.pth
   git commit -m "First working MNIST CNN"
   

3. *Crea il repo remoto su GitHub*

   * Vai su GitHub
   * Clicca su *New repository*
   * Dai un nome al repo (es. mnist_cnn)
   * Clicca *Create*

4. *Collega e invia il progetto online*
   Sostituisci tuo-utente con il tuo username GitHub:

   bash
   git remote add origin https://github.com/tuo-utente/mnist_cnn.git
   git branch -M main
   git push -u origin main

## Piccolo report – Scrivi un file README.md

Scrivi un breve README.md (max ½ pagina) con:

* *Obiettivo del progetto*
  Breve descrizione del problema affrontato (es. classificare cifre scritte a mano).

* *Sintesi del setup e del modello*
  Versione di PyTorch, numero di layer, tipo di rete usata (CNN, LSTM, ecc.).

* *Risultati*
  Metrica finale (accuracy, loss, dice score…), magari una mini-confusion matrix o esempi corretti/sbagliati.

* *Due idee di miglioramento*
  Es. usare più epoche, aumentare i dati, provare un modello pre-addestrato, ottimizzare gli iperparametri.

### 1. Sito Web con **Django**

Django è un framework full-stack che gestisce routing, database, template e sicurezza “out-of-the-box”.
**Esempi**

1. Mini-portale per scambiare appunti di corso con upload, ricerca e commenti.
2. Dashboard che visualizza report giornalieri generati da uno script Python (grafici, PDF, log).
   **Tool & skill**: Django, HTML/CSS/JS, SQLite/PostgreSQL, Docker opz., Pandas, Git.
   **Difficoltà**: ⭐⭐
   **Tempo stimato**: **20-28 h**

### 2. Web Scraper (senza IA)

Scarica pagine/API, archivia dati e notifica cambi di stato.
**Esempi**

1. Tracker prezzi Amazon: salva lo storico in SQLite e avvisa quando scende sotto soglia.
2. Bot che controlla ritardi Trenitalia e manda un SMS se il treno supera 15 min.
   **Tool & skill**: requests/httpx, BeautifulSoup4 o Playwright, Pandas, SMTP/Telegram-bot, cron/APScheduler.
   **Difficoltà**: ⭐
   **Tempo stimato**: **8-12 h**
